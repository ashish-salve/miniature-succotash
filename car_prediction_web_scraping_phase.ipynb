### Class to fetch item urls from different locations

```python
class Olx_cars():
    """to open url"""
    def __init__(self,url):
        """Takes url from instance and opens chromedriver"""
        self.url = url
        self.driver = webdriver.Chrome('chromedriver.exe')
    def open_url(self):
        """open url in chromedriver"""
#         driver = webdriver.Chrome('chromedriver.exe')
        self.driver.get(self.url)
    
    def select_loc(self):
        """select location of used car"""
        locations = EC.presence_of_all_elements_located((By.XPATH,"//a[@class='_396bV _2zs4P _1nXYG']"))
        WebDriverWait(self.driver,10).until(locations)
        loc_wbes = self.driver.find_elements_by_xpath("//a[@class='_396bV _2zs4P _1nXYG']") 
#         pdb.set_trace()
        loc_wbes[12].click()
        
    def load_all(self):
        """collects urls for all loaded items"""
        flag = True
        while flag:
            # count number of items on current webpage
            self.driver.implicitly_wait(10)
            no_of_items = len(self.driver.find_elements_by_xpath("//li[@class='EIR5N']/a"))
#             pdb.set_trace()
            # code to exit while loop and store the urls of items
            if no_of_items >= 500:
                flag = False
                # store available product urls in a list
                try:
                    urls = []
                    try:
                        WebDriverWait(self.driver,10).until(EC.presence_of_all_elements_located((By.XPATH,"//li[@class='EIR5N']/a")))
                    except TimeoutException:
                        WebDriverWait(self.driver,10).until(EC.presence_of_all_elements_located((By.XPATH,"//li[@class='EIR5N']/a")))
                    for wbe in self.driver.find_elements_by_xpath("//li[@class='EIR5N']/a"):
            #             pdb.set_trace()
                        urls.append(wbe.get_attribute('href'))
                    return urls
                except Exception as e:
                    print(e)
#             wait till we find load more button
            try:
                load_btn = EC.presence_of_element_located((By.XPATH,"//div[@class='JbJAl']/button"))
                WebDriverWait(self.driver,10).until(load_btn)
            except TimeoutException:
                load_btn = EC.presence_of_element_located((By.XPATH,"//div[@class='JbJAl']/button"))
                WebDriverWait(self.driver,20).until(load_btn)
            except NoSuchElementException:
                break
            # after , click on the button
            try:
                WebDriverWait(self.driver,10).until(EC.element_to_be_clickable((By.XPATH,"//div[@class='JbJAl']/button")))
                self.driver.find_element_by_xpath("//div[@class='JbJAl']/button").click()
            except Exception as e:
                print(e)
                print(no_of_items)
            del no_of_items
```

```python
class Getinfo():
    """fetches data from url"""
    def __init__(self,url):
        """takes url from instance and opens webdriver"""
        self.url = url
        self.driver = webdriver.Chrome('chromedriver.exe')
    def open_url(self):
        """opens url in webdriver"""
        self.driver.get(self.url)
    def fetch_model(self):
        """fetches model from current webpage"""
        model = self.driver.find_element_by_xpath("//div[@class='_35xN1']").text                
        return model
    def fetch_fuel(self):
        """fetches fuel from current webpage"""
        fuel = self.driver.find_elements_by_xpath("//div[@class='_3qDp0']")[0].text
        return fuel
    def fetch_mileage(self):
        """fetches mileage from current webpage"""
        mileage = self.driver.find_elements_by_xpath("//div[@class='_3qDp0']")[1].text
        return mileage
    def fetch_owner(self):
        """fetches owner from current webpage"""
        owner = self.driver.find_elements_by_xpath("//div[@class='_1gasz']")[0].text
        return owner
    def fetch_loc(self):
        """fetches location from current webpage"""
        loc = self.driver.find_elements_by_xpath("//div[@class='_1gasz']")[1].text
        return loc
    def fetch_price(self):
        """fetches price from current webpage"""
        price = self.driver.find_element_by_xpath("//div[@class='_3FkyT']").text
        return price
```

```python
try:
    model = []
    fuel = []
    mileage = []
    owner = []
    loc = []
    price = []
    for url in df_olx_urls['0'][5448:]:
#         time.sleep(5)
        temp_url_index = df_olx_urls['0'].index[df_olx_urls['0']==url].tolist()
        url_index = temp_url_index[0]
        get_info = Getinfo(url)
        get_info.open_url()
        
        if get_info.driver.current_url == 'https://www.olx.in/':
            get_info.driver.quit()
            url = df_olx_urls['0'][url_index]
            get_info = Getinfo(url)
            get_info.open_url()
        try:    
            if get_info.driver.find_element_by_xpath("//h4").text == "We're sorry":
                get_info.driver.refresh()
        except NoSuchElementException:
            pass
        try:    
            if get_info.driver.find_element_by_xpath("//h1").text == "502 Bad Gateway":
                get_info.driver.refresh()
        except NoSuchElementException:
            pass
        try:
            model.append(get_info.fetch_model())
            fuel.append(get_info.fetch_fuel())
            mileage.append(get_info.fetch_mileage())
            owner.append(get_info.fetch_owner())
            loc.append(get_info.fetch_loc())
            price.append(get_info.fetch_price())
            get_info.driver.quit()
        except NoSuchElementException:
            if get_info.driver.find_element_by_xpath("//img[@class='iqQ7C']"):
                get_info.driver.quit()
                continue
        del url_index
    df_olx_cars_29sept = pd.DataFrame({})
    df_olx_cars_29sept['Model'] = model
    df_olx_cars_29sept['Fuel Type'] = fuel
    df_olx_cars_29sept['Mileage'] = mileage
    df_olx_cars_29sept['Owner'] = owner
    df_olx_cars_29sept['Location'] = loc
    df_olx_cars_29sept['Price'] = price
    
    df_olx_cars_29sept.to_csv('olx_cars_final.csv',index=False)
except Exception as e:
    print(get_info.url)
    print('URL index: {}'.format(url_index))
    print(e)
    df_olx_cars_29sept = pd.DataFrame({})
    df_olx_cars_29sept['Model'] = model
    df_olx_cars_29sept['Fuel Type'] = fuel
    df_olx_cars_29sept['Mileage'] = mileage
    df_olx_cars_29sept['Owner'] = owner
    df_olx_cars_29sept['Location'] = loc
    df_olx_cars_29sept['Price'] = price
    
    df_olx_cars_29sept.to_csv('olx_cars_29sept_2.csv',index=False)
```
